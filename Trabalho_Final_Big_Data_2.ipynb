```
df = spark.read.csv("/FileStore/tables/results.csv", header=True)
```

```
df_novo = df \
    .withColumnRenamed('date','Data') \
    .withColumnRenamed('home_team','Mandante') \
    .withColumnRenamed('away_team','Visitante') \
    .withColumnRenamed('home_score','Gols_mandante') \
    .withColumnRenamed('away_score','Gols_visitante') \
    .withColumnRenamed('tournament','Campeonato') \
    .withColumnRenamed('city','Cidade') \
    .withColumnRenamed('country','Pais') \
    .withColumnRenamed('neutral','Neutro') \
    .select('Data','Mandante','Visitante','gols_mandante', 'gols_visitante','Campeonato','Cidade','Pais')
			
```

```
df_novo.registerTempTable('df_novo')
```

```
sql("select count(*) AS QTD_REGISTROS from df_novo").show()
```

```
mandante = df_novo.groupBy('Mandante').sum()
print('Mandantes ' + str(mandante.count()))
```

```
sql("select count(*) AS QTD_VITORIAS_MANDANTE from df_novo where gols_mandante > gols_visitante").show()
```

```
sql("select count(*) AS QTD_EMPATES from df_novo where gols_mandante = gols_visitante").show()
```

```
sql("select Pais, count(Pais) as Quantidade_partida  from df_novo group by Pais").show()
```

```
sql("with cte as (select Pais, count(Pais) as Quantidade_partida  from df_novo group by Pais) select * from cte order by 2 desc ").show(1)
```

```
sql("select *, (gols_visitante + gols_mandante) as Total_Gols from df_novo order by Total_Gols desc").show(1)
```

```
sql("with cte as (select Data, Mandante, Visitante, gols_mandante, gols_visitante, Campeonato, Cidade, Pais, abs(gols_mandante - gols_visitante) as Diferenca_gol from df_novo) select * from cte order by Diferenca_gol desc").show(1)
```
