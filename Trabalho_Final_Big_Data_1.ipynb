```
df = spark.read.csv("/FileStore/tables/results.csv", header=True)
```

```
df.show()
```

```
print("Quantidade de Registros: " + str(df.count()))
```

```
df_novo = df \
    .withColumnRenamed('date','Data') \
    .withColumnRenamed('home_team','Mandante') \
    .withColumnRenamed('away_team','Visitante') \
    .withColumnRenamed('home_score','Gols_mandante') \
    .withColumnRenamed('away_score','Gols_visitante') \
    .withColumnRenamed('tournament','Campeonato') \
    .withColumnRenamed('city','Cidade') \
    .withColumnRenamed('country','Pais') \
    .withColumnRenamed('neutral','Neutro') \
    .select('Data','Mandante','Visitante','gols_mandante', 'gols_visitante','Campeonato','Cidade','Pais')
```

```
df_novo.show()
```

```
numMandante = df_novo.groupBy('Mandante').sum()

print('Quantidade de times mandantes: ' + str(numMandante.count()))
```

```
numMandanteVitoria = df_novo.filter(df_novo.gols_mandante > df_novo.gols_visitante)
print('Quantidade de vitórias mandantes :' + str(numMandanteVitoria.count()))
```

```
numVisitanteVitoria = df_novo.filter(df_novo.gols_mandante < df_novo.gols_visitante)
print('Quantidade de vitórias visitantes :' + str(numVisitanteVitoria.count()))
```

```
numEmpate = df_novo.filter(df_novo.gols_mandante == df_novo.gols_visitante)
print('Quantidade de empates :' + str(numEmpate.count()))
```

```
df_novo.groupBy('Pais').count().withColumnRenamed('count', 'Quantidade de Partidas').show()
```

```
df_novo.groupBy('Pais').count().orderBy('count', ascending=False).withColumnRenamed('count', 'Quantidade de Partidas').show(1)
```

```
maiorQtdGols = df_novo.withColumn('Qtd_Gols_Por_Partida',df_novo.gols_visitante + df_novo.gols_mandante) 
maiorQtdGols.orderBy('Qtd_Gols_Por_Partida', ascending=False).show(1)
```

```
df_novo.registerTempTable('df_novo')
```

```
from  pyspark.sql.functions import abs

df_sup = df_novo.withColumn('Maior_Goleada', abs(df_novo.gols_visitante - df_novo.gols_mandante))
df_sup.orderBy('Maior_Goleada', ascending=False).show(1)
```
